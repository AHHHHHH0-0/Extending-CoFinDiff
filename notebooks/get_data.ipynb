```python
import requests
import json
import time
import os
import sys
import yfinance as yf

from dotenv import load_dotenv
from pathlib import Path

sys.path.insert(0, str(Path().resolve().parent))
from config import data_config

Path('../data/raw').mkdir(parents=True, exist_ok=True)
```

```python
"""
Get daily stock data (Alpha Vantage)
"""

# Get database URL
load_dotenv()
api_key = os.environ.get("AV_API_KEY_PREMIUM")
base_url = os.environ.get("AV_BASE_URL")

# Get data
for i, ticker in enumerate(data_config.TICKERS):
    url = f"{base_url}function={data_config.DAILY_ADJUSTED}&symbol={ticker}&outputsize=full&apikey={api_key}"
    r = requests.get(url)
    data = r.json() 

    # Save data in JSON file
    with open(f'../data/raw/daily_data_{ticker}.json', 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
    
    # Wait
    time.sleep(3)

    print(f"{i+1}/{len(data_config.TICKERS)}: Daily {ticker} data ({len(data['Time Series (Daily)'])}) saved to ../data/raw/daily_data_{ticker}.json")
```

```python
"""
Get daily IR data (Alpha Vantage)
"""

# Get database URL
load_dotenv()
api_key = os.environ.get("AV_API_KEY")
base_url = os.environ.get("AV_BASE_URL")

# Get data
url = f"{base_url}function=FEDERAL_FUNDS_RATE&interval=daily&apikey={api_key}"
r = requests.get(url)
data = r.json()

# Save to JSON file
with open('../data/raw/daily_ir.json', 'w', encoding='utf-8') as f:
    json.dump(data['data'], f, ensure_ascii=False, indent=4)

print(f"IR data saved to ../data/raw/daily_ir.json")
print(f"IR data length: {len(data['data'])}")
```

```python
"""
Get daily VIX data (Yahoo Finance)
"""

# Get data
vix = yf.Ticker("^VIX")
vix_all = vix.history(period="max")
data = vix_all.reset_index()[['Date', 'Close']]

# Date to string
data['Date'] = data['Date'].dt.strftime('%Y-%m-%d')

# Dataframe to list of dict
vix_data = data.to_dict('records')

# Save to JSON file
Path('../data/raw').mkdir(parents=True, exist_ok=True)
with open('../data/raw/daily_vix.json', 'w', encoding='utf-8') as f:
    json.dump(vix_data, f, ensure_ascii=False, indent=4)

print(f"VIX data saved to ../data/raw/daily_vix.json")
print(f"VIX data length: {len(vix_data)}")
```
